```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(42)

# Reproduce a realistic x distribution (mixture of a bulk near 0 and a sparse tail)
n = 300
x = np.concatenate([
    rng.exponential(scale=0.6, size=int(n * 0.7)),   # bulk near zero
    rng.uniform(1.5, 4.5, size=int(n * 0.3)),         # sparse high-error tail
])
x = np.clip(x, 0, 4.5)

# Ideal: unmixed error ≈ 0 for everything
y = rng.exponential(scale=0.08, size=len(x))

fig, ax = plt.subplots(figsize=(5.5, 5.5))

# Reference y=x line
lim = 4.5
ax.plot([0, lim], [0, lim], color="steelblue", linewidth=1.5)

ax.scatter(x, y, color="steelblue", alpha=0.5, s=30)

ax.set_xlim(-0.3, lim + 0.2)
ax.set_ylim(-0.3, lim + 0.2)
ax.set_xlabel("Raw absolute log2 O/E error")
ax.set_ylabel("Unmixed absolute log2 O/E error")
ax.set_title("Reference error before vs after unmixing\n(ideal)")

plt.tight_layout()
plt.show()
```
